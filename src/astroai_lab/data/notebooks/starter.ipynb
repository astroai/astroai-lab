```python
import os
import subprocess

# Apply scratch-backed caches even if the kernel missed profile hooks.
try:
    out = subprocess.check_output(["astroai-lab", "env", "export"], text=True)
    for line in out.splitlines():
        if line.startswith("export ") and "=" in line:
            body = line[len("export "):]
            k, _, v = body.partition("=")
            v = v.strip().strip("'\"")
            os.environ[k] = v
except Exception as exc:
    print("env export skipped:", exc)

!astroai-lab doctor --json | python -c "import sys,json; d=json.load(sys.stdin); print('hygiene_ok=', d.get('hygiene_ok')); print('scratch=', d.get('scratch_dir')); print('xdg_cache=', __import__('os').environ.get('XDG_CACHE_HOME'))"
```
